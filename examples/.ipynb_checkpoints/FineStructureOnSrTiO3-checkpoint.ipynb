```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from src.core.spectrum import Spectrum, Spectrumshape
from src.core.model import Model

from src.components.linear_background import LinearBG
from src.components.CLedge.hs_coreloss_edgecombined import HSCoreLossEdgeCombined
from src.components.MScatter.mscatterfft import MscatterFFT
from src.components.powerlaw import PowerLaw
from src.components.gdoslin import GDOSLin

from src.fitters.linear_fitter import LinearFitter
from src.fitters.lsqfitter import LSQFitter

from src.operators.backgroundremoval import BackgroundRemoval
```

```python
#The experimental parameters 
alpha = 0.9e-3 #convergence angle [rad]
beta = 2.675e-3 #collection angle [rad]
E0 = 300e3 #acceleration voltage [V]
```

```python
# The edges in the spectrum
elements = ['Ti', 'O']
edges = ['L', 'K']
```

# Loading the datasets

```python
filename =  r'C:\Users\DJannis\PycharmProjects\pyEELSmodel\Examples\SrTiO3_example\hl.msa'
filenamell = r'C:\Users\DJannis\PycharmProjects\pyEELSmodel\Examples\SrTiO3_example\ll.msa'
s0 = Spectrum.loadmsa(filename)
ll = Spectrum.loadmsa(filenamell)
```

```python
#the core loss spectrum
s0.plot()
```

```python
#the low loss spectrum
ll.plot()
```

# The conventional method: Powerlaw background subtraction

```python
#A list of integration windows
windows = [[454,500],[530,590]]
#The signal ranges over which needs to be fitted
ranges = [(380,450),(512,527)]

#These values are exctracted from looking at the core loss spectrum, no user interface is available
```

```python
#The background fitting procedure
intensity = np.zeros(len(windows)) #the integrated intensities will be saved in this array
for i in range(len(windows)):
    back = BackgroundRemoval(s0, signal_range=ranges[i]) #background removal object
    s0_rem = back.calculate() #calculates the background subtracted signal
    intensity[i] = s0_rem.integrate(windows[i])
    fig = back.show_fit_result(integration_range=windows[i]) #small function which shows the fitted result
    fig.axes[0].set_title(elements[i])
```

# Model based fitting

```python
#The background model, in this case it is linear
specshape = s0.get_spectrumshape()
BG = LinearBG(specshape)
```

```python
#The components of the edges
#Rather slow due to cross section computation
comp_elements = []
for elem, edge in zip(elements, edges):
    comp = HSCoreLossEdgeCombined(specshape, 1, E0, alpha, beta, elem, edge)
    comp_elements.append(comp)
```

```python
#Add fine structure on the edges
comp_fine = []
for comp in comp_elements:
    comp_fine.append(GDOSLin.gdoslin_from_edge(specshape, comp, pre_e=5, ewidth=50, degree=20))
#pre_e: is how much energy before the edge onset, the linear interpolation can be performed
#ewidth: the width of the interpolation region
#degree: The number of points used in the interpolated region
```

```python
#The low loss spectrum which is the convolutor
ll_comp = MscatterFFT(specshape, ll)
```

```python
#the list containing all the components
component_list = [BG]+comp_elements+comp_fine+[ll_comp]
```

```python
#Defining the model we are going to use to fit the spectrum
mod = Model(specshape, components=component_list) #The model object
```

```python
#Defining the fitter used which in this case is the linear fitter
fit = LinearFitter(s0, mod, method='ols') #The fitter object 
```

```python
#Performing the fit
fit.perform_fit()
fit.set_fit_values()
```

```python
fit.plot()
```

# Quantification of Ti/O

```python
# Using the intensities form the background subtracted method to quantify the Ti and O content
TiL = HSCoreLossEdgeCombined(specshape, 1, E0, alpha, beta, 'Ti', 'L')
OK = HSCoreLossEdgeCombined(specshape, 1, E0, alpha, beta, 'O', 'K')

R_th = TiL.integrate(windows[0])/OK.integrate(windows[1]) #theoretical ratio
R_abun = intensity[0]/(intensity[1]*R_th)
```

```python
# Using the model for the quantification
model_abun = fit.coeff[fit.get_param_index(comp_elements[0].parameters[0])]/fit.coeff[fit.get_param_index(comp_elements[1].parameters[0])]

#CRLB calculated error
error = fit.CRLB_ratio(comp_elements[0].parameters[0], comp_elements[1].parameters[0])
```

```python
print('Conventional method: N_Ti/N_O=' + str(np.round(R_abun,2)))
print('Model based method: N_Ti/N_O={f}+-{g}'.format(f=np.round(model_abun,3),g=np.round(error,3)))
```
